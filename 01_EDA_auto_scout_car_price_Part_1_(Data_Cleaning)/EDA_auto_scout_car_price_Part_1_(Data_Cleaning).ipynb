## <div class="alert alert-danger"><font color=black>7. <u>Droping columns that have missing values more than %80 :</u></font></div>

```python
def show_missing_values (data):
    
    '''
    
    This function plots missing values for each column by observation in the dataset.
    
    ''' 
    plt.figure(figsize=(10, 5))

    sns.displot(data=data.isnull().melt(value_name="missing"),
                y="variable",
                hue="missing",
                multiple="fill",
                height=9.25)

    plt.axvline(0.8, color="r")
    plt.show()
```

```python
show_missing_values(df)
```

```python
def df_missing (df, limit):
        
    '''
    
    This function returns records that greater than or equal to the given limit value for each column.
    
    ''' 
    
    missing_percentage = df.isnull().sum()*100 / df.shape[0]
    return missing_percentage.loc[lambda x : x >= limit]
```

```python
df_missing(df,80)
```

```python
drop_columns = df_missing(df,80).index
drop_columns
```

```python
df.drop(drop_columns, axis=1, inplace=True)
```

### <div class="alert alert-info"><font color=black>consumption :</font></div>

```python
df["consumption"].astype('str').value_counts(dropna=False)
```

```python
def cons_comb(x):
    
    if type(x)==list and len(x)==3:
        if x[0] != []: return x[0][0]
        else: return None
        
    if type(x)==list and len(x)==5:
        return x[1]
    
    if type(x)==list and len(x)==7:
        return x[1]
    
    else:
        return None
```

```python
def cons_city(x):
    
    if type(x)==list and len(x)==3:
        if x[1] != []: return x[1][0]
        else: return None
        
    if type(x)==list and len(x)==5:
        return None
    
    if type(x)==list and len(x)==7:
        return x[3]
    
    else:
        return None
```

```python
def cons_country(x):
    
    if type(x)==list and len(x)==3:
        if x[2] != []: return x[2][0]
        else: return None
        
    if type(x)==list and len(x)==5:
        return None
    
    if type(x)==list and len(x)==7:
        return x[5]
    
    else:
        return None
```

```python
df['cons_comb'] = df["consumption"].apply(cons_comb).str.extract('(\S+)').astype("float") 
df['cons_city'] = df["consumption"].apply(cons_city).str.extract('(\S+)').astype("float")
df['cons_country'] = df["consumption"].apply(cons_country).str.extract('(\S+)').astype("float")
```

```python
df["cons_city"].value_counts(dropna=False).head(10)
```

```python
df["cons_country"].value_counts(dropna=False).head(10)
```
